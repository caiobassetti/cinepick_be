```python
from diffusers import AutoPipelineForText2Image
import torch
```

```python
pipeline = AutoPipelineForText2Image.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float32, # Change to float32 if running without GPU, float 16 for GPU
    use_safetensors=True
)#.to("cuda") # If not using a GPU, remove the .to("cuda")
```

```python
prompt = "movie poster of the godfather and alice in wonderland" # The prompt can be changed here
pipeline(prompt, num_inference_steps=30).images[0] # Change the number of inference steps for variations
```
